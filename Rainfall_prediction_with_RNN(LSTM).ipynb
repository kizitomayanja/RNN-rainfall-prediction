```python
rain_dataset = {}
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
half_months = ['Jan','Feb','Mar','Apr','May','Jun']
for record in rainfall_dict:
  if record['Year'] not in rain_dataset:
    rain_dataset[record['Year']] = {}
  if record['Year'] != 2018:
    rain_dataset[record['Year']][record['Station']]={month: record[month] for month in months if month in record}
  elif record['Year'] == 2018:
    rain_dataset[record['Year']][record['Station']] = {month: record[month] for month in half_months if month in record}
  else:
    rain_dataset[record['Year']] = {record['Station']:{month: record[month] for month in months if month in record}}
```

```python
tempmax = pd.read_excel("Rain data/Maximum Temperature 2012 to 2018.xlsx")
tempmin = pd.read_excel("Rain data/Minimum Temperature 2012 to 2018.xlsx")
```

```python
# prompt: I'm converting a column of months specified as numbers into abbrevations for example 1 for Jan

month_dict = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}

# Assuming 'df' is your DataFrame and 'month_num_column' is the name of the column containing month numbers
tempmax['MONTH'] = tempmax['MONTH'].map(month_dict)
tempmin['MONTH'] = tempmin['MONTH'].map(month_dict)
```

```python
tempmax_dict = tempmax.to_dict('records')
tempmin_dict = tempmin.to_dict('records')
tempmax_dict
```

```python
def transform_dataset(dataset):
  transformed_dataset = {}
  for record in dataset:
    if record['YEAR'] not in transformed_dataset:
      transformed_dataset[record['YEAR']] = {}
      # tempmax_dataset[record['YEAR']]['Station'] = {}
    # if record['YEAR']['Station'] not in tempmax_dataset[record['YEAR']]:
    #   tempmax_dataset[record['YEAR']][record['YEAR']['Station']] = {}
    current_month = record['MONTH']
    for key in record.keys():
      if key != 'YEAR' and key != 'MONTH':
        if key not in transformed_dataset[record['YEAR']]:
          transformed_dataset[record['YEAR']][key] = {}
        # tempmax_dataset[record['YEAR']]['Station'] = key
        transformed_dataset[record['YEAR']][key][current_month] = record[key]
  return transformed_dataset
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense,Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
def transform_datetimes(dict, station):
  date_dataset = {}
  if 'date' not in date_dataset:
    date_dataset['date'] = []

  month_dict = {
      "Jan":"01",
      "Feb":"02",
      "Mar":"03",
      "Apr":"04",
      "May":"05",
      "Jun":"06",
      "Jul":"07",
      "Aug":"08",
      "Sep":"09",
      "Oct":"10",
      "Nov":"11",
      "Dec":"12"
  }

  for feature in dict:
    print(feature)
    if feature not in date_dataset:
      date_dataset[feature] = []
    for year in dict[feature]:
      # print(year)
      for month in dict[feature][year][station]:
        month_num = month_dict[month]
        datestring = str(year) + '-' + month_num + '-01'
        if datestring not in date_dataset['date']:
          date_dataset['date'].append(datestring)
        date_dataset[feature].append(dict[feature][year][station][month])
  return pd.DataFrame.from_dict(date_dataset)


```

```python
mubende_dataset_pd = transform_datetimes(dataset, 'Mubende')
mubende_dataset_pd.head(4)
```

```python
#converting the date values into datetime
train_dates=pd.to_datetime(mubende_dataset_pd['date'])
train_dates.head()
```

```python
#As we need to consider 3 attributes for prediction
cols=list(mubende_dataset_pd)[1:4]
print(cols)
```

```python
training_df=mubende_dataset_pd[cols].astype(float)
```

```python
#Plotting the last 100 values
plot_df=training_df.tail(100)
plot_df.plot.line()
```

```python
#We need to normalize the data using Standard scalar
scalar=StandardScaler()
#Considering the whole data we normalize the data
#Fit the data accordingly
scalar=scalar.fit(training_df)
#Tranforming whole data into normalized form from the fitted data
training_df_scaled=scalar.transform(training_df)
```

```python
#Lists for storing the training data considering multiple variables
#trainX for storing training data (temperature,wind speed,relative humidity)
#trainY for storing target values(rainfall)
trainX=[]
trainY=[]
```

```python
n_past=20#20 values will be considered to predict the next values
n_fut=1#for predicting the next 1 value based on previous 20 values
```

```python
#The normalized values are
training_df_scaled
```

```python
#Adding the values to trainX and trainY lists
for i in range(n_past,len(training_df_scaled)-n_fut+1):
    trainX.append(training_df_scaled[i-n_past:i,0:training_df_scaled.shape[1]])
    trainY.append(training_df_scaled[i+n_fut-1:i+n_fut,0])
```

```python
#Converting into numpy arrays
import numpy as np
trainX,trainY=np.array(trainX),np.array(trainY)
```

```python
print(trainX)
```

```python
#Based on 20 previous values next value has to be predicted in a slidding manner
print(trainY)
```

```python
print('trainX shape == {}'.format(trainX.shape))#Knowing the shape
#(364,20,4) means there are 364windows(groups) of 20*4
#In our case, we have 58 windows(groups) of 20(records) * 3(features)
```

```python
print('trainY shape == {}'.format(trainY.shape))
#There are 58 values from previous values
```

```python
model = Sequential()
model.add(LSTM(128,activation='relu',input_shape=(trainX.shape[1],trainX.shape[2]),return_sequences=True))
model.add(LSTM(64,activation='relu',return_sequences=True))
model.add(LSTM(32,activation='relu',return_sequences=True))
model.add(LSTM(16,activation='relu',return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))
```

```python
#optimizer used is adam with loss function mse(Mean Squared Error)
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
#Fitting the model
history=model.fit(trainX,trainY,epochs=100,batch_size=5,validation_split=0.2,verbose=1)
```
